```python
#solver：svd：奇异值分解，不计算协方差矩阵，因此适用于具有大量特征的数据集；lsqr：最小二乘法，可与正则化一起使用；eigen：特征值分解，可与正则化一起使用，适用于特征数不多的数据集。 
#shrinkage：正则化参数，可以增强LDA分类的泛化能力。若是只用LDA降维，可以不用。参数类型为string或float(0~1之间)，默认是None
#priors：类别权重，用于分类问题时。若是只用LDA降维，可以不用。参数类型为array，大小为类别个数n_classes。
#n_components：它的作用是指定LDA降维后的特征数(也就是降维后的维度)。参数类型为int，默认为None，1≤n_components<n_classes-1。若是不做降维，则不设置，用默认即可。
```

```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
from sklearn.model_selection import train_test_split
import numpy as np
def rlda(X,y,params):
    lda=LinearDiscriminantAnalysis(**params)
    result1=lda.fit_transform(X,y)
    names=["lda_components_"+str(i) for i in range(result1.shape[1])]
    result2=pd.DataFrame(result1,columns=names)
    df_new=pd.concat([X,result2],axis=1)
    print(lda.explained_variance_ratio_)
    return df_new
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
X=pd.DataFrame(data=iris.data,columns=iris.feature_names)
y=data=iris.target
params={
    #"solver" :,
    #"shrinkage" :,
    #"priors":,
    "n_components" :2
    #"store_covariance" :,
    #"tol" :
}
rlda(X,y,params)
```

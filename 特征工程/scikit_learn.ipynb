```python
import matplotlib.pyplot as plt
from sklearn import datasets
data,label=datasets.make_blobs(n_samples=100,n_features=3,centers=6)
plt.scatter(data[:,0],data[:,1],c=label)
plt.show()
```

```python
from sklearn import preprocessing
data = [[0, 0], [1, 0], [-1, 1], [1, 2]]
scalerstd=preprocessing.StandardScaler().fit(data)
print(scalerstd.mean_)
print(scalerstd.var_)
```

```python
#上一题的scaler标准化现有数据
scalerstd.transform(data)
```

```python
#最大最小规范化
sc=preprocessing.MinMaxScaler(feature_range=(0,1)).fit(data)
sc.transform(data)
```

```python
#L2正则化
sc=preprocessing.Normalizer(norm="l2").fit(data)
sc.transform(data)
```

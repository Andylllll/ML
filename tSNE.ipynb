```python
#n_components：嵌入空间的维度 (default: 2)
#perplexity：复杂度与其他流形学习算法中使用的最近邻数有关。较大的数据集通常需要较大的困惑。考虑选择一个介于5和50之间的值。不同的值会导致不同的结果。
#early_exaggeration：(默认值:12.0)控制原始空间中的自然集群在嵌入式空间中的紧密程度，以及它们之间的空间大小。对于较大的值，在嵌入的空间中，自然簇之间的空间会更大。同样，这个参数的选择不是很关键。如果在初始优化过程中代价函数增加，则早期夸大因子或学习率可能过高。
#learning_rate ： t-SNE的学习率通常在[10.0，1000.0]范围内。 如果学习率太高，数据可能看起来像“球”点与其最近的邻居大约等距。 如果
#n_iter: int，可选(默认值:1000)优化的最大迭代次数。至少250个
#n_iter_without_progress:int，可选的(默认值:300)在我们终止优化之前没有进展的最大迭代数，在250个最初的迭代之后使用，并带有早期的夸张效果。注意，进度每50次迭代才检查一次，所以这个值四舍五入为下一个50的倍数。
#min_grad_norm:浮动，可选(默认值:1e-7)如果梯度范数低于这个阈值，优化将停止。
#random_state 
# n_jobs
```

```python
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd
def rtsne(df,params):
    tsne=TSNE(**params)
    result1=tsne.fit_transform(df)
    names=["tsne_components_"+str(i) for i in range(result1.shape[1])]
    result2=pd.DataFrame(result1,columns=names)
    df_new=pd.concat([df,result2],axis=1)
    
    return df_new
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
params={
    "n_components" :3
   # "perplexity" :,
   # "early_exaggeration"  :,
   # "learning_rate"  :,
   # "n_iter"  :,
   # "n_iter_without_progress"  :,
    #"min_grad_norm"  :,
   # "metric"  :,
   #"init"  :,
   # "verbose"  :,
   #"random_state"  :,
   #"method"  :,
   #"angle" :,
   #"n_jobs" :
}
rtsne(df,params)
```

```python
#catboost:完全对称二叉树,XGboost是一层一层地建立节点，lightGBM一个一个地建立节点，而Catboost总是使用完全二叉树
#支持分类特征，不需要对分类特征进行预处理（LabelEncoding 或 OneHotEncoding）
#iterations:创建树的最大数量
#learning_rate:学习率  automatically or 0.03
#one_hot_max_size=2: 对于所有具有多个不同值小于或等于给定参数值的特性，使用one-hot编码
#eval_metric：评估标准（RMSE Logloss MAE CrossEntropy Recall Precision F1 Accuracy AUC R2）
#depth：树的深度（推荐值：1-10） 
#subsample:数据行的采样率
#random_strength:选择树的结构时用于对拆分评分的随机量，使用此参数可以避免过度拟合
#min_data_in_leaf：一个叶子中训练样本的最小数量
#colsample_bylevel, colsample_bytree, colsaml2_leaf_reg(ple_bynode ：分别表示各个层、各棵树、各个节点的列采样率
#rsm：对于那些具有几百个特性的数据集，rsm参数加快了训练的速度，通常对训练的质量不会有影响。另外，不建议为只有少量（10-20）特征的数据集更改rsm参数的默认值
#border_count：此参数定义了每个特征的分割数。默认情况下，如果在CPU上执行训练，它的值设置为254，如果在GPU上执行训练，则设置为128

```

```python
import numpy as np
import pandas as pd
import catboost.datasets
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Imputer,MinMaxScaler
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
import catboost as cb
iris = datasets.load_iris()
X = iris.data
y = iris.target
scaler=MinMaxScaler(feature_range=(0,1))
iris = datasets.load_iris()
X=scaler.fit_transform(X)
```

```python
def GridSearch(model,params,X,y):
    cbr=GridSearchCV(model,params,scoring='neg_mean_squared_error',cv=5)
    cbr.fit(X,y)
    print('GridSearchCV_best_score:',cbr.best_score_)
    print('GridSearchCV_best_params：',cbr.best_params_) 
    print('GridSearchCV_best_model：',cbr.best_estimator_) 
```

```python
params={
        'l2_leaf_reg':3,
        'bagging_temperature':1,
        'random_strength':1,
        'rsm':1,
        'loss_function' :'RMSE',
        'eval_metric':'RMSE'
     
}
model=catboost.CatBoostRegressor(**params)
change_params={
                'n_estimators' :range(10,100,10),
                'learning_rate':np.arange(0.03,0.1,0.02),
                'depth':range(1,10,2)
                }
GridSearch(model,change_params,X,y)
```

```python
#RandomSearchCV:并不是所有的参数都尝试，具有随机性
def RandomSearch(model,params,X,y):
    cbr=RandomizedSearchCV(model,params,scoring='neg_mean_squared_error',cv=5)
    cbr.fit(X,y)
    print('RandomizedSearchCV_best_score:',cbr.best_score_)
    print('RandomizedSearchCV_best_params：',cbr.best_params_) 
    print('RandomizedSearchCV_best_model：',cbr.best_estimator_) 
```

```python
#贝叶斯优化
from sklearn.model_selection import train_test_split,cross_val_score	
from bayes_opt import BayesianOptimization
def BayesianSearch(model,params):
    num_iters=30
    init_points=5
    bayes= BayesianOptimization(model, params)
    bayes.maximize(init_points=init_points,n_iter=num_iters)
    params = bayes.res
    index = []
    for i in bayes.res:
        index.append(i['target'])
    max_index = index.index(max(index))

    print(params[max_index])
    return max_index
        
```

```python

def evaluate(n_estimators, learning_rate, depth):
    param={
        'l2_leaf_reg':3,
        'bagging_temperature':1,
        'random_strength':1,
        'rsm':1,
        'loss_function' :'RMSE',
        'eval_metric':'RMSE'
    }#固定参数
    param['n_estimators'] = int(n_estimators)
    param['depth'] = int(depth)
    param['learning_rate'] = float(learning_rate)
    
    #超参数
    cvs=cross_val_score(cb.CatBoostRegressor(**param),X,y,scoring='neg_mean_squared_error',cv=5).mean()
    return cvs


    
```

```python
change_params={
                'n_estimators':(10,100),
                'learning_rate':(0.01,1),
                'depth':(1,10)
                }
```

```python
BayesianSearch(evaluate,change_params)
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Imputer,MinMaxScaler
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
import xgboost as xgb
from lightgbm.sklearn import LGBMRegressor

bo = datasets.load_boston()
X = bo.data
y = bo.target
scaler=MinMaxScaler(feature_range=(0,1))
iris = datasets.load_iris()
X=scaler.fit_transform(X)


```

```python
#GridSearchCV（就是穷举）
#xgboost参数：
#n_estimator:迭代次数（生成树的个数）
#min_child_weight[默认1]：决定最小叶子节点权重和，用于避免过拟合
#max_depth（default：6）：树的最大深度，一般取值3-6
#gamma：控制节点分裂
#subsample和colsample_bytree：
#reg_alpha和reg_lambda：正则化选项
#learning_rate：学习率
def GridSearch(model,params,X,y):
    xgbr=GridSearchCV(model,params,scoring='neg_mean_squared_error',cv=5)
    xgbr.fit(X,y)
    print('GridSearchCV_best_score:',xgbr.best_score_)
    print('GridSearchCV_best_params：',xgbr.best_params_) 
    print('GridSearchCV_best_model：',xgbr.best_estimator_) 
```

```python
param = {'n_estimators': 400,
        'learning_rate': 0.1,
         'max_depth': 5, 
         'min_child_weight': 1,
         'seed': 0,
         'subsample': 0.8,
         'colsample_bytree': 0.8, 
         'gamma': 0, 
         'reg_alpha': 0, 
         'reg_lambda': 1}

model= xgb.XGBRegressor(**param)
change_params = {'n_estimators': range(10, 200, 20),
     #            'min_child_weight': range(3, 10, 2),
      #           'max_depth': range(3, 15, 3),
      #           'gamma': np.arange(0.1,0.9,0.2),
      #           'subsample': np.arange(0.5, 1.0, 0.1),
      #        'colsample_bytree': np.arange(0.5, 1.0, 0.1),
      #         'reg_lambda': np.arange(0.1, 1.0, 0.4),
       #         'reg_alpha': np.arange(0.1, 1.0, 0.4),
        #      'learning_rate': np.arange(0.01, 0.2, 0.05)
                }
GridSearch(model,change_params ,X,y)
```

```python
#RandomSearchCV:并不是所有的参数都尝试，具有随机性
def RandomSearch(model,params,X,y):
    xgbr=RandomizedSearchCV(model,params,scoring='neg_mean_squared_error',cv=5)
    xgbr.fit(X,y)
    print('RandomizedSearchCV_best_score:',xgbr.best_score_)
    print('RandomizedSearchCV_best_params：',xgbr.best_params_) 
    print('RandomizedSearchCV_best_model：',xgbr.best_estimator_) 
```

```python
param = {'n_estimators': 20,
        'learning_rate': 0.1,
         'max_depth': 5, 
         'min_child_weight': 1,
         'seed': 0,
         'subsample': 0.8,
         'colsample_bytree': 0.8, 
         'gamma': 0, 
         'reg_alpha': 0, 
         'reg_lambda': 1}
model= xgb.XGBRegressor(**param)
change_params = {'n_estimators': range(20, 200, 20),
  #               'min_child_weight': range(3, 10, 2),
#              'max_depth': range(5, 15, 3),
 #                'gamma': np.arange(0.1,0.9,0.2),
  #               'subsample': np.arange(0.5, 1.0, 0.1),
   #              'colsample_bytree': np.arange(0.5, 1.0, 0.1),
    #             'reg_lambda': np.arange(0.1, 1.0, 0.4),
     #            'reg_alpha': np.arange(0.1, 1.0, 0.4),
      #          'learning_rate': np.arange(0.01, 0.2, 0.05)
                }
RandomSearch(model,change_params ,X,y)
```

```python
#贝叶斯优化

from sklearn.model_selection import train_test_split,cross_val_score	
from bayes_opt import BayesianOptimization
def BayesianSearch(model,params):
    num_iters=20
    init_points=5
    bayes= BayesianOptimization(model, params)
    bayes.maximize(init_points=init_points,n_iter=num_iters)
    params = bayes.res
    index = []
    for i in bayes.res:
        index.append(i['target'])
    max_index = index.index(max(index))

    print(params[max_index])
    return max_index
        
```

```python

def evaluate(learning_rate,max_depth,min_child_weight,subsample,colsample_bytree,gamma,reg_alpha,reg_lambda):
    param={
        'seed': 0,
        'n_estimators':20
    }#固定参数
    param['min_child_weight']=int(min_child_weight)
    param['max_depth']=int(max_depth)
    param['gamma']=float(gamma)
    param['subsample']=float(subsample)
    param['colsample_bytree']=float(colsample_bytree)
    param['reg_lambda']=float(reg_lambda)
    param['reg_alpha']=float(reg_alpha)
    param['learning_rate']=float(learning_rate)
    #超参数
    
    #交叉检验
    cvs=cross_val_score(xgb.XGBRegressor(**param),X,y,scoring='neg_mean_squared_error',cv=5).mean()
    return cvs


    
```

```python
change_params={
             'min_child_weight': (3,10),
              'max_depth': (5, 15),
                'gamma': (0.1,1),
                 'subsample':(0.5, 1.0),
               'colsample_bytree': (0.5, 1.0),
               'reg_lambda':(0.1, 1.0),
               'reg_alpha': (0.1, 1.0),
             'learning_rate': (0.01, 1)
}
```

```python
BayesianSearch(evaluate,change_params)
```

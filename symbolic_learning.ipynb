```python
import pandas as pd
from gplearn.genetic import SymbolicTransformer
from sklearn.model_selection import train_test_split

def rsl(X,y,params):
    gp = SymbolicTransformer(**params)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
    gp.fit(X_train, y_train)
    names = ["symbolic_component_" + str(i) for i in range(gp.n_components)]
    result = pd.DataFrame(gp.transform(X),columns=names)
    df_new = pd.concat([X,result],axis=1)
    return df_new
```

```python

from sklearn.datasets import load_boston
boston=load_boston()
X=pd.DataFrame(data=boston.data,columns=boston.feature_names)
X=X[['CRIM', "ZN"]]
y=boston.target
params={
       "population_size":100,
 #       "hall_of_fame":
       "n_components":12
  #      "generations ":
    #    "tournament_size"
  #      "stopping_criteria "
    #    " const_range"
   #     "init_depth" 
   #      "init_method"
   #      "function_set":('add', 'sub', 'mul', 'div'),
   #     "transformer" 
    #    " metric"
     #    "parsimony_coefficient"
      #   "p_crossover"
       #  "p_subtree_mutation"
        # "p_hoist_mutation"
         #"p_point_mutation"
       #   "p_point_replace"
       #   "max_samples"
      #    "feature_names"
      #    "warm_start"
     #     "low_memory"
       #   "n_jobs"
     #     "verbose"
     #    "random_state"
}
rsl(X,y,params)
```
